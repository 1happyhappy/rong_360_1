```python
import pandas as pd
import datetime as dt
import numpy as np
import os
import gc
import matplotlib as  mpl
from matplotlib  import pyplot as plt
mpl.rcParams[u'font.sans-serif'] = ['simhei']
mpl.rcParams['axes.unicode_minus'] = False
%matplotlib inline
import seaborn as sns
import joblib
```

## 1.加载所有日期数据和label数据

```python
input_dir = '../preprocess_data_new/'
train_ax_date = joblib.load(input_dir + 'train_ax_date.lz4')
valid_date = joblib.load(input_dir + 'valid_date.lz4')
test_date = joblib.load(input_dir + 'test_date.lz4')

train_y = joblib.load(input_dir + 'train_y_33465.lz4')
```

## 2.训练集各个月份样本数分布

```python

plt.figure(figsize=(20,20))
plt.subplot(221)
(train_ax_date.loan_month.value_counts().sort_index()/100000).plot.bar()
plt.title(s = 'train集各个月份样本数分占比',fontdict = {'fontsize': 25})
plt.ylabel(s = '数量占比', fontdict = {'fontsize': 20})
plt.xlabel(s = '月份', fontdict = {'fontsize': 20})
plt.yticks(fontsize=20)
plt.xticks(rotation=0, fontsize=20)

plt.subplot(222)
(valid_date.loan_month.value_counts().sort_index()/20000).plot.bar()
plt.title(s = 'valid集各个月份样本数分占比',fontdict = {'fontsize': 25})
plt.ylabel(s = '数量占比', fontdict = {'fontsize': 20})
plt.xlabel(s = '月份', fontdict = {'fontsize': 20})
plt.yticks(fontsize=20)
plt.xticks(rotation=0, fontsize=20)

plt.subplot(223)
(test_date.loan_month.value_counts().sort_index()/20000).plot.bar()
plt.title(s = 'test集各个月份样本数分占比',fontdict = {'fontsize': 25})
plt.ylabel(s = '数量占比', fontdict = {'fontsize': 20})
plt.xlabel(s = '月份', fontdict = {'fontsize': 20})
plt.yticks(fontsize=20)
plt.xticks(rotation=0, fontsize=20)
```

## 3.一个月中每一天的数据量分布

```python
plt.figure(figsize=(16,10))
plt.subplot(221)
(train_ax_date.loan_day.value_counts().sort_index()/100000).plot(kind='bar')
plt.xticks(rotation=0)
plt.title('train集数据与日期的关系')
plt.ylabel(s = '数量占比', fontdict = {'fontsize': 10})
plt.xlabel(s = '天', fontdict = {'fontsize': 10})

plt.subplot(222)
(valid_date.loan_day.value_counts().sort_index()/20000).plot(kind='bar')
plt.xticks(rotation=0)
plt.title('valid集数据与日期的关系')
plt.ylabel(s = '数量占比', fontdict = {'fontsize': 10})
plt.xlabel(s = '天', fontdict = {'fontsize': 10})

plt.subplot(223)
(test_date.loan_day.value_counts().sort_index()/20000).plot(kind='bar')
plt.xticks(rotation=0)
plt.title('test集数据与日期的关系')
plt.ylabel(s = '数量占比', fontdict = {'fontsize': 10})
plt.xlabel(s = '天', fontdict = {'fontsize': 10})
```

## 4.查看loan_week_day 与 逾期率之间的关系

```python
# 逾期率随时间起伏变大
label_df = train_y[['label']]
week_day_df = train_ax_date.loc[:33465,'loan_week_day']
df = pd.concat([week_day_df, label_df],axis=1)
df.groupby(by=['loan_week_day']).mean().plot()
plt.hlines(label_df['label'].mean(),0,7,colors='r')
plt.ylabel(s = '逾期率', fontdict = {'fontsize': 10})
plt.xlabel(s = 'loan_week_day', fontdict = {'fontsize': 10})
plt.title('逾期率的变化幅度随着loan_week_day的增大而增大')
```
