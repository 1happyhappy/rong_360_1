```python
def show_uniq(data):
    '''
    描述：
        展示unique num 的走势
    '''
    uniq = [[col,data[col].unique().shape[0]] for col in data.columns]
    uniq_sort = sorted(uniq,key=lambda x: x[1],reverse=False)
    uniq_num = [i[1] for i in uniq_sort]
    
    plt.figure(figsize=(20,10))
    plt.subplot(121)
    plt.plot(uniq_num)
    plt.title(s = 'UniqueNum 按升序排列', fontdict={'fontsize':20})
    plt.ylabel(s = 'UniqueNum', fontdict={'fontsize':15})
    plt.xlabel(s = '特征', fontdict={'fontsize':15})

    plt.subplot(122)
    plt.title(s = 'lg(UniqueNum) 按升序排列', fontdict={'fontsize':20})
    plt.plot(np.log10(uniq_num))
    plt.ylabel(s = 'lg(UniqueNum)', fontdict={'fontsize':15})
    plt.xlabel(s = '特征', fontdict={'fontsize':15})
```

## 2.通过观察UniqueNum的走势，寻找合适的cut_point

```python
# 通过观察UniqueNum的走势，可知阶梯递增曲线小时在1和2之间，
# 离散值与连续值的划分点在100，1000之间
show_uniq(train_ax)
```
